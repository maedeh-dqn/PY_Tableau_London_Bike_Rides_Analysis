```python
#Extract the file from the downloaded zip file
zipfile_name = './LondonBikeRides/london-bike-sharing-dataset.zip'
with zipfile.ZipFile(zipfile_name, 'r') as file:
    file.extractall()
```

```python
#Read in the CSV file as a Pandas dataframe
lbr = pd.read_csv('london_merged.csv')
```

```python
#Specifying the column names we want to use throughout the analysis
new_cols_dict = {
    'timestamp': 'time',
    'cnt':'count',
    't1':'temp_real_C',
    't2':'temp_feels_like_C',
    'hum':'humidity_percent',
    'wind_speed':'wind_speed_kph',
    'weather_code':'weather',
    'is_holiday':'is_holiday',
    'is_weekend':'is_weekend',
    'season':'season'
}

#Renaming the columns based on the new_cols_dict
lbr.rename(new_cols_dict, axis=1, inplace=True)
```

```python
#Changing the humidity values to percentage (a value between 0 and 1)
lbr.humidity_percent = lbr.humidity_percent / 100
```

```python
lbr
```

```python
#Create season dictionary to map the integer 0-3 to the actual written values
season_dict = {
    '0.0':'spring',
    '1.0':'summer',
    '2.0':'autumn',
    '3.0':'winter'
}
#Changing the season column data type to string
lbr.season = lbr.season.astype(str)

#Map the values 0-3 to the actual written seasons
lbr.season = lbr.season.map(season_dict)
```

```python
#Create weather dictionary to map the integers to the actual written values
weather_dict = {
    '1.0':'Clear',
    '2.0':'Scattered clouds',
    '3.0':'Broken clouds',
    '4.0':'Cloudy',
    '7.0':'Rain',
    '10.0':'Rain with thunderstorm',
    '26.0':'Snowfall'
}

#Changing the weather column data type to string
lbr.weather = lbr.weather.astype(str)

#Map the values to the actual written weathers
lbr.weather = lbr.weather.map(weather_dict)
```
